```python
##  halfyear, 최소 물질의 양 : initial
## 현재 물질의 양 current, 년도 : year

initial = 100
halfyear= 10
year = 0
current = initial
while current > initial / 10.0:
      year += halfyear
      current -= current / 2.0
      print(f"{year}년 {current}g")
print(f"{year}년 후에 잔류량은 {current}g 이다.")
```

```python
## 종이접기 면적 구하기
## 종이를 접으면 면적이 1/2  일 때 1/100일 때 접이 횟수

inital =1
current = initial
cnt= 0
while 1:
  current /=2
  cnt +=1
  print(f"{cnt}회 : {current}")
  if current < initial / 100:
       break
  print(f"{cnt}회 : {current}")
```

```python
## 3.종일ㄹ 접으면 높이가 2배가 된다고 할 때
## 에베레스트 산의 높이 8800m 높아지려면 횟수
## 종이두께는 0.1mm

initial = 0.1 #mm
current = initial/1000
cnt=0
while 1:
  current *= 2
  cnt += 1
  print(f"{cnt} 회 : {current}m")
  if current >8800:
    break
    print(f"{cnt} 회 : {current}m")
```
